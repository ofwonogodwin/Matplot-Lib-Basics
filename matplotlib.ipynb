Matplotlib For Data Visualization


In Normal Life, Humans process Images More Faster than Text and That Is The Biggest Importance of Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#Drawing Line Graphs

```python
#Below is a Python list Of Values containing Days of The week and their recorded Temperatures.
# We are going To use Them To Visualize The Simple Data In the List
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
#The Code Below is for labeling the X-axis, Y-axis, and the Title Respectively
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
#The ".plot" is the Plotting Object That Actually Brings Out The Visualization You See
plt.plot(x,y,color = 'green',marker ='o')
#Other options for markers are [^ for triangles, D for diamonds,* for stars,o for solid Circles]
#You can Do the Color Configurations with colors of Your Choice
```

```python
#Drawing a Line Graph With Multiple Values now
days = [1,2,3,4,5,6,7]
max_t = [50,52,52,45,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,43,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
#Here we are Going to Use the Concept of Legends
#lagends are Normally Used as Keys To Know what Line Represents what Temperature
#When dealing with Legends, after Ploting, you need to and a label as seen below
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label ='Min')
plt.plot(days,avg_t,label = 'Avg')
# The loc Concept your see Below Is to Position The key(Upper Right, Upper Left,Best,Lower Right and Many More Positions)
plt.legend(loc = 'best',shadow = True)
#The Grid Brings Kind of the Graph lines
plt.grid()
```

#Dealing with Histograms

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, color ='green',bins =3,rwidth =0.95)
```

```python
#Plotting Two Cases On One Histogram.
blood_sugar_men =[100,180,150,95]
blood_sugar_women = [113,90,84,105,120]
```

```python
plt.ylabel('Number Of People')
plt.xlabel('Blood_Sugar Levels')
plt.title('Blood Sugar Levels Of Men And Women')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150,180],rwidth = 0.95,color=['green','orange'],label =['Men','Women'])
plt.legend(loc = 'best')
```

#Dealing with Pie Charts

```python
item = ['Rent','Food','Car','Water','Other']
cost = ['300','500','500','50','600']
```

```python
plt.pie(cost,labels = item,radius =1.0,explode = [0,0,0,0,0.11])
plt.title('Expenditure For the Month')
plt.show()
```
